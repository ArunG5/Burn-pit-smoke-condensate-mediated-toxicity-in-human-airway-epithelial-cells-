```r
head(nemenyis_df)
```

```r
# adding a col to denote whether a nemenyis had a significant p adjust value, log FC, or both
nemenyis_df$log2FC = as.numeric(nemenyis_df$log2FC)
nemenyis_df$P.Adj = as.numeric(nemenyis_df$P.Adj)

# add a column of NAs
nemenyis_df$diffexpressed <- "No Significance"

# if p adjust is significant only
nemenyis_df$diffexpressed[nemenyis_df$P.Adj < 0.2] <- "P Adjust Only"

# if log2FC > log2(1.3) and P.Adj < 0.2, set as "UP" 
nemenyis_df$diffexpressed[abs(nemenyis_df$log2FC) > abs(log2(1.3)) & nemenyis_df$P.Adj < 0.2] <- "P Adjust and log2 Fold Change"

# making into a factor
nemenyis_df$diffexpressed = factor(nemenyis_df$diffexpressed, levels = c("No Significance", "P Adjust Only", 
                                                                       "P Adjust and log2 Fold Change"))

# putting cols into a factor
nemenyis_df$Condensate = factor(nemenyis_df$Condensate)
nemenyis_df$Comparison = factor(nemenyis_df$Comparison)

# viewing df
head(nemenyis_df)
```
